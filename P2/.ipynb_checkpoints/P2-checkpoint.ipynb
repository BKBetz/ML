#### AND Test
We gaan eerst de AND port testen. We geven de Perceptron willekeurige waardes en gaan kijken hoelang het duurt voordat de Perceptron op het goeie antwoord komt met behulp van de learning rule.

```python
# random perceptron
p1 = Perceptron([20, 38], 11.5)
print(p1.activationfunc([0,0]))
```

Het bovenste voorbeeld had 0 moeten zijn maar is een 1. We maken nu een functie voor de learning rule en passen dit toe

```python
"""
    The loop given below is a loop that makes it possible to run any random perceptron 
    through the learning rule and compeletly fix it (at least it is suppose to). It keeps looping through all possible inputs
    and is keeping track of a score. If the score is the same amount as the maximum inputs (in this case 4) it stops.
    The loop needs to change based on the amount of inputs and the target that you want.
    
    An AND port has the targets [0,0,0,1] so the loop below is based on that. But a XOR port has the targets [0,1,1,0].
    if that is the case you need to change the target values of the port. Keep this in mind if you want to test different things
"""


def AND_lr():
    while p1.score != 4:
        p1.score = 0
        for x in range(0,2):
            for y in range(0,2):
                if [x,y] == [0,0]:
                    p1.update([x,y], 0, 0.8)
                elif [x,y] == [0,1]:
                    p1.update([x,y], 0, 0.8)
                elif [x,y] == [1,0]:
                    p1.update([x,y], 0, 0.8)
                elif [x,y] == [1,1]:
                    p1.update([x,y], 1, 0.8)
AND_lr()    
```

Nu tonen we het resultaat

```python
print(p1.activationfunc([0,0]))
print(p1.activationfunc([0,1]))
print(p1.activationfunc([1,0]))
print(p1.activationfunc([1,1]))

print(p1)
```

### XOR Test
We doen nu dezelfde test als met de AND port, alleen nu met de XOR

```python
p2 = Perceptron([0.1, 0.5], 0.8)
p2.activationfunc([0,0])
```

Het bovenste voorbeeld had moeten eindigen met een 0, maar is dus 1. Echter Is het een XOR niet linear scheidbaar en kan het niet opgelost worden met maar 1 perceptron. Als je de code hieronder runt zal die nooit stoppen (test het maar)

```python
"""
    Same loop as the AND port. except different targets (as said before)
"""
def XOR_lr():
    while p2.score != 4:
        p2.score = 0
        for x in range(0,2):
            for y in range(0,2):
                if [x,y] == [0,0]:
                    p2.update([x,y], 0, 0.8)
                elif [x,y] == [0,1]:
                    p2.update([x,y], 1, 0.8)
                elif [x,y] == [1,0]:
                    p2.update([x,y], 1, 0.8)
                elif [x,y] == [1,1]:
                    p2.update([x,y], 0, 0.8)
                    
        print("gaat em niet worden")
        
XOR_lr()  
```

### IRIS Dataset
We gaan nu proberen de learning rule toe te passen op de iris dataset. Hier gebruiken we een random seed voor.

```python
random.seed(1760236)
iris = load_iris()
iris
```

We maken nu een perceptron met 4 random weights en een random bias

```python
p3 = Perceptron([random.randint(0,9),random.randint(0,9),random.randint(0,9),random.randint(0,9)], random.randint(0,9))
```

Nu pakken we de juiste data uit de iris dataset en gebruiken passen we de learning rule op de perceptron toe. We gebruiken ook een random learning rate

```python
data, targets = iris['data'], iris['target']
correct = 0

for x in range(0,len(data)):
    for y in range(0, 10):
        p3.update(data[x], targets[x], random.randint(0,1))
    print(p3.activationfunc(data[x]), targets[x])
    if p3.activationfunc(data[x]) == targets[x]:
        correct += 1

print(correct)
    
    
```
